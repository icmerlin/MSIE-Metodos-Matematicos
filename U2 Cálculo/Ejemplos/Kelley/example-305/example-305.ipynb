```python
# Step 1: Define the vector field components
x, y = var('x y')
F = (y, -x)

# Step 2: Plot the vector field
vector_field_plot = plot_vector_field(F, (x, -3, 3), (y, -3, 3), color='blue')

# Step 3: Define the general solution (from previous steps)
t = var('t')
C1, C2 = var('C1 C2')
x_sol = C1 * cos(t) + C2 * sin(t)
y_sol = -C1 * sin(t) + C2 * cos(t)

# Step 4: Plot sample solution curves
# Example initial conditions for different trajectories
curves = []
for C1_value, C2_value in [(-2, 0), (1, 1), (0, -2)]:
    x_curve = x_sol.subs(C1=C1_value, C2=C2_value)
    y_curve = y_sol.subs(C1=C1_value, C2=C2_value)
    curve_plot = parametric_plot((x_curve, y_curve), (t, 0, 2*pi), color='red')
    curves.append(curve_plot)
    curves.append(point((x_curve.subs(t=0), y_curve.subs(t=0)), color='green', size=30))

# Combine vector field and sample solution curves
final_plot = vector_field_plot + sum(curves)
final_plot.show()
```
